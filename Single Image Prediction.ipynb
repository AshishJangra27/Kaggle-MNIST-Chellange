```python
import cv2 as cv
import numpy as np
from keras.models import load_model
import matplotlib.pyplot as plt

model = load_model('CNN.hdf5')
```

```python
def prediction():
    
    frame     =  cv.imread("Test/three.png")
    im     =  cv.imread("Test/three.png",0)
    im     =  cv.resize(im,(28,28))
    
    im = im/255.0
    im = 1. - im
    im = np.reshape(im,(1,28,28,1))

    title = 'It is '+str(np.argmax(model.predict(im)))

    im = np.reshape(im,(28,28))
    
    plt.imshow(frame)
    plt.title(title)
    plt.show()
    
prediction()
```
